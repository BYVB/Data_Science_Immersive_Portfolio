# Project 1 - Data Exploration, summary statistics for SAT Scores. 

##### Data Summary

This dataset describes the rate of test completion and average test scores for SAT math and verbal sections in each state in the US. It includes a 51st value for Washington DC. Data source: College Board https://research.collegeboard.org/programs/sat/data. Data year: 2001.

##### Data dictionary:

State (String): Two letter state identifiers following postal/ISO codes. Includes Washington DC as distinct entry.

Rate (Int): Percent of test completion per state as of 2001. 

Verbal (Int): Average score from 1-800 on the verbal section of the SAT as of 2001.
    
Math (Int): Average score from 1-800 on the math section of the SAT as of 2001.




## Load the data.

```python
a = []
with open("../data/sat_scores.csv", 'r') as f:
    for ii, line in enumerate(f):
        a.append((line.strip().split(',')))
    

print a
```

##### Remove labels

```python
###make new data object with only lists after the first
l= a[1:]

###update this data object in place to remove the last list in the list of lists
l.pop()
print l
```

##### Save object of State names.

```python
state_names = []
for row in l:
    state_names.append(row[0])

print state_names


    
```

##### Amend types

```python
#check item type from a random sub-list
[type(item) for item in l[5]]

```

```python
#make new data object, with ints instead of strings for cols 1:4
for row in l:
    for x in range(1,4):
        row[x] = int(row[x])
```

```python
##double check that all worked...
[type(item) for item in l[5]]
```

##### Create a dictionary for each State and its respective values.

```python
###Solution version with one dict for all values, states as keys

sat_state_dict = {}
for row in l:
    sat_state_dict[row[0]] = row[1:] ##cycle through indices except for 0
print sat_state_dict
```

##### Unique dictionary for each score type.


```python
### Three dicts: one for rate, verbal, math with states as keys and their scores as values

verbal = {}
for row in l:
    verbal[row[0]]=row[2]

math = {}
for row in l:
    math[row[0]]=row[3]

rate = {}
for row in l:
    rate[row[0]]=row[1]

print verbal
```

##### Final dictionary with all numerical values

```python
## one dict which has rate, verbal, and math as keys and a list of the states' scores as values

values_dict = {}

##reference original data for header titles to populate keys
titles = a[0]

values_dict[titles[1]] = rate.values()
values_dict[titles[2]] = verbal.values()
values_dict[titles[3]] = math.values()

print values_dict
```

## Descriptive analytics

```python
## call max function on each column of the main data matrix, print them with descriptive strings

print "Rate max: ",  max(row[1] for row in l), ", Rate min: ", min(row[1] for row in l)
print "Verbal max: ", max(row[2] for row in l),  ", Verbal min: ", min(row[2] for row in l)
print "Math max: ", max(row[3] for row in l),  ", Math min: ", min(row[3] for row in l)
```

```python
#make a function for the average
def get_avg(x):
    return (sum(x)*1.0)/len(x)
    
```

```python
#make a function for the stdeviation
def make_stdv(x):
    variance = [(abs(i-get_avg(x))**2) for i in x]
    return (sum(variance)/len(x))**0.5
```

```python
###print values for the numerical fields
rate_dev = make_stdv(rate.values())
verbal_dev = make_stdv(verbal.values())
math_dev = make_stdv(math.values())

print "Standard deviation for Rate: ", rate_dev
print "Standard deviation for Verbal: " , verbal_dev
print "Standard deviation for Math: ", math_dev
```

##### Though normal would be nice...

A typical distribution for a large sample of data is a Normal or Gaussian distribution. This means that approximately 65% of the sample population will fall in the range between one standard deviation above and below the mean. Almost all of the sample size (~95%) will fall between +-2x standard deviation, and that 99.7% will be in a range of +-3x standard deviations from the mean. 

##### ... it's not always in the cards

Our data does not follow a normal distribution upon visual inspection of the histograms for rate values or verbal values. Math values are closer to a normal distribution, however show noticeable variation. The basic math statements to test the validity of these claims programmatically are detailed below. Ultimately, since the datum represent aggregations (state averages) this is not a large enough sample size for distribution to be measured effectively. 

```python
math_avg = get_avg(math.values())
range_65 = str(math_avg - math_dev) + " - " + str(math_avg + math_dev)

### Print statement about math's actual values. 

print "Math's average is ", math_avg, " with a std deviation of ", math_dev
print "We'd expect 65 percent of the population to fall between ", range_65 
```
